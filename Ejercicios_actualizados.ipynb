```python

import pandas as pd
import numpy as np


### CARGA DE DATOS
colombia = {
    'Amazonas': ['Leticia', 'Puerto Nariño'],
    'Antioquia': ['Medellín', 'Envigado', 'Bello', 'Itagüí', 'Rionegro', 'Santa Fe de Antioquia'],
    'Atlántico': ['Barranquilla', 'Soledad', 'Malambo', 'Puerto Colombia'],
    'Bolívar': ['Cartagena de Indias', 'Magangué', 'Turbaco', 'El Carmen de Bolívar'],
    'Boyacá': ['Tunja', 'Duitama', 'Sogamoso', 'Villa de Leyva'],
    'Caldas': ['Manizales', 'Villamaría', 'Chinchiná', 'La Dorada'],
    'Caquetá': ['Florencia', 'San Vicente del Caguán'],
    'Cauca': ['Popayán', 'Santander de Quilichao', 'Puerto Tejada'],
    'Cesar': ['Valledupar', 'Aguachica', 'Codazzi'],
    'Chocó': ['Quibdó', 'Istmina', 'Condoto'],
    'Córdoba': ['Montería', 'Lorica', 'Sahagún'],
    'Cundinamarca': ['Bogotá', 'Zipaquirá', 'Chía', 'Girardot', 'Soacha', 'Facatativá'],
    'Guaviare': ['San José del Guaviare'],
    'Huila': ['Neiva', 'Pitalito', 'Garzón'],
    'La Guajira': ['Riohacha', 'Maicao', 'Uribia'],
    'Magdalena': ['Santa Marta', 'Ciénaga', 'El Banco'],
    'Meta': ['Villavicencio', 'Acacías', 'Granada'],
    'Nariño': ['Pasto', 'Tumaco', 'Ipiales'],
    'Norte de Santander': ['Cúcuta', 'Ocaña', 'Pamplona'],
    'Putumayo': ['Mocoa', 'Puerto Asís'],
    'Quindío': ['Armenia', 'Calarcá', 'Montenegro', 'La Tebaida'],
    'Risaralda': ['Pereira', 'Dosquebradas', 'Santa Rosa de Cabal'],
    'San Andrés y Providencia': ['San Andrés', 'Providencia'],
    'Santander': ['Bucaramanga', 'Floridablanca', 'Girón', 'Barrancabermeja'],
    'Sucre': ['Sincelejo', 'Corozal'],
    'Tolima': ['Ibagué', 'Espinal', 'Melgar'],
    'Valle del Cauca': ['Cali', 'Palmira', 'Buenaventura', 'Tuluá', 'Cartago', 'Buga']
}

tipos_energia = ['Hidráulica', 'Solar', 'Eólica', 'Biomasa', 'Geotérmica', 'Carbón', 'Gas Natural', 'Diesel']

n = 100    # Número de proyectos
### CREAR ESTRUCTURAS DE ALMACENAMIENTO
municipios = []
energias = []
generaciones = []
consumos = []
poblaciones = []
balances = []
consumos_per_cap = []
estados = []
proyectos = []

### CREAR DATOS SINTÉTICOS
nombres_deptos = list(colombia.keys())
departamentos = np.random.choice(nombres_deptos, size=n)

contador = 1
for departamento in departamentos:
    municipio = np.random.choice(colombia[departamento])
    municipios.append(municipio)
    energia = np.random.choice(tipos_energia)
    energias.append(energia)
    generacion = np.random.randint(3_000, 10_000)
    generaciones.append(generacion)
    consumo = np.random.randint(2_000, 8_000)
    consumos.append(consumo)
    poblacion = np.random.randint(5_000, 20_000)
    poblaciones.append(poblacion)
    balance = generacion - consumo
    balances.append(balance)
    cons_per_cap = consumo / poblacion
    consumos_per_cap.append(cons_per_cap)
    proyecto = f'Proyecto {contador}'
    proyectos.append(proyecto)
    contador = contador + 1
    estado = 'Sostenible' if balance >= 0 else 'Crítico'
    estados.append(estado)

### CREAR LA ESTRRCUTURA FINAL DE ALMACENAMIENTO - DICCIONARIO

data_cobertura = {
    'Proyecto' : proyectos,
    'Departamento' : departamentos,
    'Municipio' : municipios,
    'Tipo de Energia' : energias,
    'Generacion' : generaciones,
    'Consumo' : consumos,
    'Poblacion' : poblaciones,
    'Balance' : balances,
    'Consumo per Capita' : consumos_per_cap,
    'Estado' : estados
}

df = pd.DataFrame(data_cobertura)

variables = df.shape[1] # despliegue para el dashboard tener en cuenta
registros= df.shape [0]


print(f'''

forma del Dataset:
numero de variables:  {variables}
numero de registros:  {registros}
''')

display(df.head(5))  #5 primeros registros
display(df.tail(3))  # 5 ultimos registros
```
